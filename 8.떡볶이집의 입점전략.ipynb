```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/KDT/데이터분석/데이터/소상공인시장진흥공단_상가(상권)정보_서울_202303.csv')
df.info()
```

```python
shop = ['엽기떡볶이', '신전떡볶이', '죠스떡볶이', '국대떡볶이','우리할매떡볶이','감탄떡볶이','불타는여고24시떡볶이']
```

```python
# 파리바게트(파리바게뜨) 데이터 필터링
# 상가업소번호, 상호명, 경도, 위도
# contains(): 특정 문자열 포함 여부를 True, False로 반환

data = df['상호명'].str.contains(r'파리바게뜨|파리바게트') # r은 원시 문자열
df_paris = df.loc[data, ['상가업소번호', '상호명', '경도', '위도']].copy()
df_paris
```

```python
df_paris = df_paris.set_axis(['업소번호', '업소상호명', '업소경도', '업소위도'], axis=1).reset_index(drop=True)
df_paris
```

```python
# 떡볶이 프랜차이즈 데이터 필터링
df_shop = df.copy()
# df_shop[df_shop['상호명'].isin(shop)]
# extract() : 특정 문자열을 포함하고 있으면 그 문자열 반환하고, 포함하지 않으면 NaN 반환
df_shop['상호명'] = df_shop['상호명'].str.extract(r'({})'.format('|'.join(shop)))[0]
df_shop = df_shop.dropna(subset=['상호명']).iloc[:, [0,1,14, 37, 38]].reset_index(drop =True)
df_shop
```

```python
# 곱집합
df1 = pd.DataFrame(['A', 'B'])
df2 = pd.DataFrame(['가','나','다'])
df1.merge(df2, how = 'cross')
```

```python
# 두 지점의 위도와 경로를 입력하면 거리를 구해주는 라이브러리
# 하버사인 공식
!pip install haversine
```

```python
from haversine import haversine
```

```python
seoul = [37.541, 126.986]
paris = [48.8567, 2.3508]

print(haversine(seoul, paris, unit = 'km'))
print(haversine(seoul, paris, unit = 'm'))
```

```python
df_cross = df_shop.merge(df_paris, how = 'cross')
df_cross
```

```python
df_cross['거리'] = df_cross.apply(lambda x: haversine([x['위도'], x['경도']], [x['업소위도'], x['업소경도']], unit='m'), axis=1)
df_cross
```

```python
# 개별 떡볶이 매장과 파리바게트와의 거리
df_dis = df_cross.groupby(['상가업소번호','상호명'])['거리'].min().reset_index()
df_dis
```

```python
# 각 프랜차이즈 별 파리바게트와의 평균 거리
df_dis.groupby('상호명')['거리'].mean()
```

```python
# agg(): 다중집계작업을 간단하게 해주는 함수
df_dis.groupby('상호명')['거리'].agg(['mean', 'count'])
```

```python
# 파리바게트와 거리가 50m 이하인 프랜차이즈별 매장의 평균거리와 매장 개수
dis = df_dis['거리'] <= 50
df_dis[dis].groupby('상호명')['거리'].agg(['mean', 'count'])
```

```python
# 위 결과를 시각화 시킬 수 있는 여러가지 도구를 이용해보자.
!pip install pandasecharts
```

```python
# 거리를 입력하면 프랜차이즈 별 파리바게트와의 평균거리와 매장 개수를 출력하는 함수
def distance(x):
    dis = df_dis['거리'] <= x
    return df_dis[dis].groupby('상호명')['거리'].agg(['mean', 'count'])

distance(100)
```
